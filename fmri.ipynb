```python
#Question
#Plot using sns.relpot,line plot,box plot on fmri data set
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',hue='event',size='region',data=fmri)
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.boxplot(x='timepoint',y='signal',data=fmri)
```
